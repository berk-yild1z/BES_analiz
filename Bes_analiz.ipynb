```python
fon_tercih=pd.read_excel("Data.xlsx",sheet_name="Fon Grubu Tercihi")
fon_tercih["Tarih"]=pd.to_datetime(fon_tercih["Tarih"],format="%Y%m")
fon_tercih.set_index("Tarih",inplace=True)
fon_tercih
```

```python
fig, ax = plt.subplots(1, 5, figsize=(20, 5))  # Boyutu daha uygun hale getirdim

for i, column in enumerate(fon_tercih.columns):
    ax[i].plot(fon_tercih.index, fon_tercih[column], marker='o', label=column)
    ax[i].set_xlabel('Tarih')
    ax[i].set_ylabel('Değer')
    ax[i].set_title(f'{column}')
    ax[i].get_yaxis().set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
    ax[i].legend()
    ax[i].grid(True)
    ax[i].tick_params(axis='x', rotation=45)
fig.suptitle('Yatırım Terchilerinin Zaman İçinde Değişimi', fontsize=16)
plt.tight_layout()  # Tüm alt grafikler için yerleşimi ayarla
plt.show() 
```

```python
model=pd.read_excel("Data.xlsx",sheet_name="Model")
model.drop(columns=["Participants", "Population", "High School", "University", "Master", "Doctorate", "Education", "25-29", "30-34", "35-39", "40-44", "45-49", "50-54", "55-59", "60-64", "Age","State Province"], inplace=True)
model
```

```python
year_index_int = model['Year'].astype(int)
model.set_index(['State Name', year_index_int], inplace=True)

```

```python
model.index.names = ['Şehir Adı', 'Yıl']
model.drop(columns="Year",inplace=True)
model
```

```python
model.rename(columns={"Participants/Population":"Katılımcı_Nüfus_Oranı","GDP Per Capita": "Kişi_Başına_Düşen_GSYH","House Sales":"Konut_Satışı","Education/Age":"Eğitim_Yaş_Oranı"},inplace=True)
```

```python
model_1 = PanelOLS.from_formula('Katılımcı_Nüfus_Oranı ~ Kişi_Başına_Düşen_GSYH + Konut_Satışı + Eğitim_Yaş_Oranı + EntityEffects',model)
results = model_1.fit()
print(results.summary)
```

```python
corr_matrix = model[['Katılımcı_Nüfus_Oranı', 'Kişi_Başına_Düşen_GSYH', 'Konut_Satışı', 'Eğitim_Yaş_Oranı']].corr()
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Korelasyon Matrisi')
plt.show()
```
